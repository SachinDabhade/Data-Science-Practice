# Day 16 - Pandas Merge() function

```python
d1 = pd.DataFrame({'ID': [1,2,3,4,5],
                  'Name': ['Sachin', 'Vaibhav', 'Dhiraj', 'Nikhil', 'Shubham']})
```

```python
d1
```

```python
d2 = pd.DataFrame({'ID': [1,2,3,4,5],
                  'Roll': ['Data Scie1ntist', 'M.Sc Computer', 'Mechanical Diploma', 'Computer Hardware', 'Mechanical Sandwitch']})
```

```python
d2
```

```python
pd.merge(d1, d2, on='ID')
```

```python
pd.merge(d2, d1)
```

```python
# Merging 2 df through how parameter
d1 = pd.DataFrame({'ID': [1,2,3,4,6],
                  'Name': ['Sachin', 'Vaibhav', 'Dhiraj', 'Nikhil', 'Shubham']})
d2 = pd.DataFrame({'ID': [1,2,3,4,5],
                  'Roll': ['Data Scie1ntist', 'M.Sc Computer', 'Mechanical Diploma', 'Computer Hardware', 'Mechanical Sandwitch']})
```

```python
pd.merge(d1, d2, on='ID', how='left')
```

```python
 pd.merge(d1, d2, on="ID", how='right')
```

```python
pd.merge(d1, d2, on='ID', how='outer')
```

```python
pd.merge(d1, d2, how='inner')
```

```python
# Indicator is default false, we are making it ture
pd.merge(d1, d2, how='outer', indicator=True)
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4], 'Name': ['A', 'B', 'C', 'D']})
df1
```

```python
df2 = pd.DataFrame({'ID': [5,6,7,8], 'Name': ['A', 'B', 'C', 'D']})
df2
```

```python
# If no column is similar then we have to use the left and right index i.e true
pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4], 'Name': ['A', 'B', 'C', 'D']})
df2 = pd.DataFrame({'ID': [1,2,3,4], 'Name': ['A', 'B', 'C', 'D']})
```

```python
pd.merge(df1, df2, on='ID')
```

```python
# Now, to solving the problem on how column name can be same
pd.merge(df1, df2, on='ID', suffixes=('_Higher', '_Lower'))
```

# Day 19 - Python join() function

```python
df1 = pd.DataFrame({'A': [1,2,3,4], 'B': [10, 20 ,30 ,40]})
df2 = pd.DataFrame({'C': [1,2,3,4], 'D': [50, 60, 70 ,80]})
```

```python
display(df1, df2)
```

```python
# Both must have same index
df1.join(df2)
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4], 'B': [10, 20 ,30 ,40]})
df2 = pd.DataFrame({'C': [1,2,3], 'D': [50, 60, 70]})
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='right')
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4], 'B': [10, 20 ,30 ,40]})
df2 = pd.DataFrame({'A': [1,2,3], 'B': [50, 60, 70]})
```

```python
df1.join(df2, lsuffix='_1')
```

```python
df1.join(df2, rsuffix='_1')
```
